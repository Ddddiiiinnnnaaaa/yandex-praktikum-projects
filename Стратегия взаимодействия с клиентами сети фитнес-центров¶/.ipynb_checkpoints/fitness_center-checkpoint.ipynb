## Модель прогнозирования оттока клиентов

Построим модель бинарной классификации клиентов, где целевой признак - факт оттока клиентов в следующем месяце.

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df.drop(columns=['churn'])
y = df['churn']

# разделяем модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# зададим алгоритм логистической регрессии для нашей модели
model = LogisticRegression(solver='liblinear')

# обучим модель
model.fit(X_train,y_train)

# воспользуемся уже обученной моделью, чтобы сделать прогнозы
predictions = model.predict(X_test)
probabilities = model.predict_proba(X_test)[:,1]
```

```python
# выведим все изученные метрики для полученного прогноза
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
print('F1: {:.2f}\n'.format(f1_score(y_test, predictions)))
print('ROC_AUC: {:.2f}'.format(roc_auc_score(y_test, probabilities)))
```

```python
# зададим алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
# обучим модель случайного леса
rf_model.fit(X_train, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test)
rf_probabilities = rf_model.predict_proba(X_test)[:,1]
```

```python
# выведим все изученные метрики для полученного прогноза
print('Accuracy: {:.2f}' .format(accuracy_score(y_test, rf_predictions)))
print('Precision: {:.2f}' .format(precision_score(y_test, rf_predictions)))
print('Recall: {:.2f}' .format(recall_score(y_test, rf_predictions)))
print('F1: {:.2f}' .format(f1_score(y_test, rf_predictions)))
print('ROC_AUC: {:.2f}'.format(roc_auc_score(y_test, rf_predictions)))
```

Сравнивая модели логистической регрессии и случайного леса  по метрикам на валидационной выборке лучше себя показала модель логистической регрессии.

## Кластеризация клиентов

Отложим в сторону столбец с оттоком и проведем кластеризацию клиентов

```python
# обязательная стандартизация данных перед работой с алгоритмами
sc = StandardScaler()
X_sc = sc.fit_transform(X)
# построим матрицу расстояний на стандартизированной матрице признаков
linked = linkage(X_sc, method = 'ward') 
```

```python
# визуализируем дендограмму
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Дендограмма')
plt.show() 
```

По дендограмме проведу линию в том месте, где начинаются 4 различных цвета я бы задала количество кластеров равное 15.

```python
# обучим модель на основании алгоритма K-Means
km = KMeans(n_clusters=5, random_state=0) 
# задаём число кластеров, равное 5, и фиксируем значение random_state для воспроизводимости результата
labels = km.fit_predict(X_sc)
# применяем алгоритм к данным и формируем вектор кластеров 
# сохраняем метки кластера в поле нашего датасета
df['cluster'] = labels
# выводим статистику по средним значениям наших признаков по кластеру
df.groupby(by='cluster').mean()
```

По делению на кластеры можно увидеть, что есть разделение по месту жительства, по количеству длительности абонемента, участие в партнерской программе, времени с момента первого обращения и средней частоте посещений.

```python
# построим распределение признаков для каждого кластера
columns_category = [
    'gender',
    'near_location',
    'partner',
    'promo_friends',
    'contract_period',
    'group_visits',
    'age',
    'avg_additional_charges_total',
    'avg_class_frequency_total']
for column in columns_category:
    
    plt.figure(figsize=(12,4))
    sns.kdeplot(x=column, hue = 'cluster', data = df, multiple='stack')
    plt.title(column)
    plt.show()
```

По этим графикам можно сказать, что в 0 кластер больше всего попало клиентов, которые живет или работают в районе расположения фитнес центра, купили абонимент на 12 месяцев, давно являются клиентами и часто посещают его. Пользуются доп.услугами и ходят на групповые занятия.

```python
# посчитаем долю оттока для каждого кластера
df.groupby('cluster')['churn'].sum()/df['churn'].sum()*100
```

Самая большая доля оттока из 3 кластера, затем идет 2 кластер. Меньше всего отток из 0 кластера.

Итак, по делению на кластеры можно увидеть, что есть разделение по месту жительства, по количеству длительности абонемента, времени с  момента первого обращения и средней частоте посещений.
В кластер с самой маленькой долей оттока  больше всего попало клиентов, которые живет или работают в районе расположения фитнес центра, купили абонемент на 12 месяцев, давно являются клиентами и часто посещают его. Пользуются доп.услугами и ходят на групповые занятия.
