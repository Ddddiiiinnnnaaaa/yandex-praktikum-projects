Заполним пропуски в столбцах ceiling_height, living_area, kitchen_area. 

Будем использовать медиану.
Чтобы данные были более точны, буду использовать количество комнат и площадь квартиры

```python
# Заполним возможные данные ceiling_height
# Будем использовать медиану
# Чтобы данные были более точны, буду использовать количество комнат и площадь квартиры

import math #нужен для проверки NaN

# функция, что будет вычислять медиану для конкретного набора по количеству комнат и площади квартиры
# в случае, если среднее не может быть вычислено (не найдется в основных значениях группы параметров)
# берем среднее по одному - сперва по количеству комнат, потом уже по площади квартиры
def calc_median_ceiling_height(df, rooms, total_area):
    # ориентируемся на оба параметра
    median = df[(df['rooms'] == rooms) & (df['total_area'] == total_area)]['ceiling_height'].median()
    # но т.к. возможна уникальная ситуация, где нет совпадения для двух параметров, предусмотрим рассчет по 1му
    if math.isnan(median):
        median = df[df['rooms'] == rooms]['ceiling_height'].median()
        if math.isnan(median):
            median = df[df['total_area'] == total_area]['ceiling_height'].median()    
    return median

# идем по всем уровням образования и типам сотрудников, что встречаются только для пустых значений
# и проставляем только для них медиану, соответствующую их группе
for rooms in df[df['ceiling_height'].isnull()]['rooms'].unique():
    for total_area in df[df['ceiling_height'].isnull()]['total_area'].unique():
        median = calc_median_ceiling_height(df, rooms, total_area)
        searchRow = (df['rooms'] == rooms) & (df['total_area'] == total_area) & (df['ceiling_height'].isnull())
        df.loc[searchRow, 'ceiling_height'] = df.loc[searchRow, 'ceiling_height'].fillna(median)
```

```python
#  тоже самое проделаем для столбца living_area
def calc_median_living_area(df, rooms, total_area):
    median = df[(df['rooms'] == rooms) & (df['total_area'] == total_area)]['living_area'].median()
    if math.isnan(median):
        median = df[df['rooms'] == rooms]['living_area'].median()
        if math.isnan(median):
            median = df[df['total_area'] == total_area]['living_area'].median()    
    return median
for rooms in df[df['living_area'].isnull()]['rooms'].unique():
    for total_area in df[df['living_area'].isnull()]['total_area'].unique():
        median = calc_median_living_area(df, rooms, total_area)
        searchRow = (df['rooms'] == rooms) & (df['total_area'] == total_area) & (df['living_area'].isnull())
        df.loc[searchRow, 'living_area'] = df.loc[searchRow, 'living_area'].fillna(median)
```

```python
#  тоже самое проделаем для столбца kitchen_area
def calc_median_kitchen_area(df, rooms, total_area):
    median = df[(df['rooms'] == rooms) & (df['total_area'] == total_area)]['kitchen_area'].median()
    if math.isnan(median):
        median = df[df['rooms'] == rooms]['kitchen_area'].median()
        if math.isnan(median):
            median = df[df['total_area'] == total_area]['kitchen_area'].median()    
    return median
for rooms in df[df['kitchen_area'].isnull()]['rooms'].unique():
    for total_area in df[df['kitchen_area'].isnull()]['total_area'].unique():
        median = calc_median_kitchen_area(df, rooms, total_area)
        searchRow = (df['rooms'] == rooms) & (df['total_area'] == total_area) & (df['kitchen_area'].isnull())
        df.loc[searchRow, 'kitchen_area'] = df.loc[searchRow, 'kitchen_area'].fillna(median)
# проверяем, что все прошло хорошо и не осталось ли пустых значений
df.isnull().sum()
```

```python
# узнаем какое значение должно быть вместо пропуска в столбце ceiling_height
df[(df['rooms'] == rooms) & (df['total_area'] == total_area)]['ceiling_height'].median()
```

Добавим в таблицу столбец в котором будет отображаться этаж - первый, последний, другой. Напишем функцию для определения этажа и применим её к столбцу floor.

```python
floors_total = df['floors_total']

def find_floor(row):
    floor = row['floor']
    if floor == 1:
        return('первый')
    elif floor == row['floors_total']:
        return('последний')
    return('другой')

df['floor_type'] =df.apply(find_floor, axis=1)
# метод apply() применяет функцию к столбцу датафрейма
df.head()
#выведем 5 первых строк на экран, что бы проверить, что все новые строки появились в таблице
```
